```python
def sort_dataset():
    filenames = os.listdir(DIR)
    data_ret = None
    label_ret = None
    for filename in filenames:
        label = np.array([0])
        if filename.lower().split('_')[3].count("fall"):
            label = np.array([1])
        min_sample_count = 270 # truncate the samples
        names = ["TimeStamp", "Sample No", "X-Axis", "Y-Axis", "Z-Axis", "Sensor Type", "Sensor ID"]
        data = pd.read_csv(DIR+filename, names=names, comment='%', sep=";")
        acc_data = data[data["Sensor Type"] == 0]
        acc_wrist_data = acc_data[acc_data["Sensor ID"] == 3]
        acc_wrist_data["total acc"] = np.sqrt(acc_wrist_data["X-Axis"]**2 + 
                                        acc_wrist_data["Y-Axis"]**2 +
                                        acc_wrist_data["Z-Axis"]**2)
        value = np.reshape(acc_wrist_data["total acc"].values, (1,-1))
        if value.shape[1] < 270:
            continue
        else:
            if data_ret is None:
                label_ret = label
                data_ret = value[:, :min_sample_count]
            else:
                print(value.shape)
                label_ret = np.vstack((label_ret, label))
                data_ret = np.vstack((data_ret, value[:, :min_sample_count]))
    return data_ret, label_ret

sorted_data, sorted_label = sort_dataset()
```
